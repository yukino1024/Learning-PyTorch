```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_data = pd.read_csv('standard.csv')
train_data = all_data[all_data['dataset'] == 'train']
train_x = train_data['hours_std']
train_y = train_data['y']

train_x = torch.tensor(train_x,dtype=torch.float32,device='cuda').unsqueeze(1)
train_y = torch.tensor(train_y,dtype=torch.float32,device='cuda').unsqueeze(1)
```

```python
class LogisticModle(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.linear = torch.nn.Linear(1,1,bias=True)

    def forward(self,x):
        return torch.sigmoid(self.linear(x))
    
modle = LogisticModle().to('cuda')
```

```python
criteion = torch.nn.BCELoss()

optimizer = torch.optim.ASGD(modle.parameters(),lr=0.01)
```

```python
for epoch in range(1,1000):
    y_hat = modle(train_x)

    loss = criteion(y_hat,train_y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(f"epoch:{epoch};  a:{modle.linear.weight.item()};  b:{modle.linear.bias.item()};  loss:{loss.item()}")
```
